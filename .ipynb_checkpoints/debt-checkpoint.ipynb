# Impact of Lockdown on Immunity

Consider a seasonal SIRS model with a period of no circulation.


```python
%matplotlib inline
import numpy as np
import scipy as sp
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
def sirs_dx(t,x,betafun,A,gamma,omega,V):
    yr = t/365.0
    beta = betafun(yr)
    if (x[0] > 0):
        vv = V
    else:
        vv = 0.
    dx = np.zeros(3)
    dx[0] = -beta*(x[0])*(x[1]) + omega*x[2] - vv
    dx[1] = beta*(x[0])*(x[1]) - gamma*x[1]
    dx[2] = gamma*x[1] - omega*x[2] + vv
    return dx
```

```python
beta0 = 0.55
A = 0.3
gamma = 0.2
R0 = beta0/gamma
omega = 0.003
V = 0.
i0 = 0.0001
ypre = 16
ylock = 1
ypost = 3
tmax = (ypre + ylock + ypost)*365.0
```

```python
def mybf(yr):
    if (yr <= ypre) or (yr >= (ypre + ylock)):
        return (beta0 + A*np.sin(2.0*np.pi*yr))
    else:
        return 0.5*gamma
```

```python
odefun = lambda t,x: sirs_dx(t,x,mybf,A,gamma,omega,V)
sol = integrate.solve_ivp(odefun,
                          (0.0,tmax), #t_eval
                          np.array([1.-i0, i0, 0.0]), #initial conditions
                          'RK45',
                          atol=1e-12,
                          rtol=1e-12,
                         )

tt = (sol.t/365.0) - 13
ii = 100.*(sol.y[1]).T
ss = 100.*(sol.y[0]).T

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.plot(tt,ii, c='indianred')
plt.xlabel('Years')
plt.ylabel('Community Prevalence (%)')
plt.xlim([0.,7])
plt.ylim([0.,1.1*np.max(ii)])
plt.subplot(1,2,2)
plt.plot(tt,ss, c='dodgerblue')
plt.xlabel('Years')
plt.ylabel('Community Susceptibility (%)')
plt.xlim([0.,7])
plt.ylim([0.,100])
plt.tight_layout()
plt.savefig('./rsv_stylised.pdf')
```
